```python
# IMPORT DEPENDENCIES
```

```python
import pandas as pd
import numpy as np 
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import Dense
```

```python
# IMPORT DATA
```

```python
Churn_dataframe = pd.read_csv('Churn.csv')
```

```python
# ANALYZING DATAFRAME 
```

```python
# head of dataframe
Churn_dataframe.head()
```

```python
# info about dataframe
Churn_dataframe.info()
```

```python
# statiscal summary of dataframe
Churn_dataframe.describe()
```

```python
# shape of dataframe
Churn_dataframe.shape
```

```python
# FEATURES (X) AND TARGET VARIABLE(y) FOR PREDICTIONS
```

```python
X = pd.get_dummies(Churn_dataframe.drop(['Churn', 'Customer ID'], axis=1))
y = Churn_dataframe['Churn'].apply(lambda x: 1 if x=='Yes' else 0)
```

```python
# SPLIT DATA INTO TRAINING AND TESTING SETS
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
# check head for 'y_train'
y_train.head()
```

```python
# checking 'y' datatype and shape
print(type(y_train))
print(y_train.shape)
```

```python
# SCALE INPUT FEATURES
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# BUILD MODEL 
```

```python
model = Sequential()
model.add(Dense(units=12, activation='relu', input_dim=X_train.shape[1]))
model.add(Dense(units=64, activation='relu'))
model.add(Dense(units=1, activation='sigmoid'))
```

```python
# COMPILE MODEL
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics='accuracy')
```

```python
# FIT AND PREDICT MODEL
```

```python
model.fit(X_train, y_train, epochs=200, batch_size=12)
```

```python
# converting data type
X_train = X_train.astype('float32')
```

```python
# converting data type
y_train = y_train.astype('int32')
```

```python
model.fit(X_train, y_train, epochs=200, batch_size=12)
```

```python
# PREDICT MODEL
```

```python
# converting data type
X_test = X_test.astype('float32')
# making prediction
y_hat = model.predict(X_test)
y_hat = [0 if val < 0.5 else 1 for val in y_hat]
```

```python
# CHECKING ACCUARACY
```

```python
accuracy_score(y_test, y_hat)
```
